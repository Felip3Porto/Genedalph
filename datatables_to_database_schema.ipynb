```python
GSEA_df = pd.DataFrame(columns=["reports_id", "Term","es","nes","pval","fdr","geneset_size","matched_size","path_to_plot","PreRanking_prerank_id"])

list_of_GOs= []
i = -1
ree = 0
for e in index_good: 
    temp_df = pd.read_csv(e)
    dirname = e.replace('gseapy.prerank.gene_sets.report.csv', '')
    #walk the directory for matching file names
    list_terms = temp_df['Term'].tolist()
    i = i + 1
    
    for file in os.listdir(dirname):
        test = file.replace('.prerank.png', '')
        ###fuckers had "_" in their filename
        check = test.replace('_', ':')
        
        for index, row in temp_df.iterrows():
                if (row[0] == check):
                    term = check
                    path = dirname + test + '.prerank.png'
                    es = row[1]
                    nes = row[2]
                    pval = row[3]
                    fdr = row[4]
                    geneset_size = row[5]
                    matched_size = row[6]
                    prerank_id = index_good_id[i]
                    ree = ree + 1 
                    
                    GSEA_df = GSEA_df.append({
                        "reports_id": ree,
                        "Term": term,
                        "es": es,
                        "nes": nes,
                        "pval": pval,
                        "fdr": fdr,
                        "geneset_size": geneset_size,
                        "matched_size": matched_size,
                        "path_to_plot": path,
                        "PreRanking_prerank_id": prerank_id
                    }, ignore_index=True)
                    
    
    
```
